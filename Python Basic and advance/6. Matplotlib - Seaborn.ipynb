# Continous Plot (features that are int and float)

```python
# importing library
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.arange(0,10)
y = np.arange(10,20)
```

```python
plt.scatter(x,y,c='g')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Graph")

```

```python
y = x*x
plt.plot(x , y , 'r*',linestyle="dashed",linewidth = 2 , markersize = 12 )
```

## Subplots

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r-*')
plt.subplot(2,2,2)
plt.plot(x,y,'g-*')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'y-*')
```

```python
plt.subplot(2,1,1)
x = np.arange(0,20)
y = np.sin(x)
plt.plot(x , y , "r-*")

plt.subplot(2,1,2)
x = np.arange(10,20)
y = np.sin(x)
plt.plot(x , y , "g-*")
```

## Barplots 

```python
x = np.arange(0,10)
y = np.arange(10,20)

x1 = [3,6,3]
y1 = [9,8,2]

plt.bar(x,y,color= "g")
plt.bar(x1,y1,color = "b")
```

## Histograms

```python
data = np.array([29,45,43,65,24,23,1,6,5,2,24,53,3,2,1])
plt.hist(data)
```

# Seaborn

#### In this problem statement we have to analyse the tip for the future customer on basis of other independent features
> tip is dependent feature

> Rest are independent feature

```python
import seaborn as sns 
```

```python
df = sns.load_dataset("tips")
df.head()
```

## Correlation and Heatmap

- Correlations are used to give the relationship between the features
- Correlation can be applied for only int and floating numbers , strings arent included while calculating correlation

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

## joint plot

- this plot is mainly used when  , there are 2 feautures 

```python
sns.jointplot(x = "total_bill" , y= "tip" , data = df)
```

```python
sns.jointplot(x = "total_bill" , y= "tip" , data = df , kind="hex")
```

```python
sns.jointplot(x = "total_bill" , y= "tip" , data = df , kind = "reg")
```

## Pair plot

- this plot is used when there are more than 2 features
- on int and float feautures , pair plot can be applied

```python
sns.pairplot(data = df , kind="reg")
```

```python
sns.pairplot(df,hue="sex")
```

```python
sns.pairplot(df,hue = "smoker")
```

## Dist plot
- this is mainly used to tell the density of a certain column

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False)
```

# Categorical plot(featuers that are categorical )

### 1.Count plot

```python
sns.countplot("tip", data = df)
```

```python
sns.countplot('day',data = df)
```

```python
sns.countplot(y = 'sex' , data=df)
```

### 2. bar plot

```python
sns.barplot(x = 'total_bill' , y ="sex" , data= df)
```

```python
sns.barplot(x = 'total_bill' , y ="smoker" , data= df)
```

### 3.Box plot

```python
sns.boxplot(x = 'total_bill' , y = 'day' , hue='sex', data=df)
```

### 4.Violin Plot

```python
sns.violinplot(x = 'day' , y = 'tip' ,data=df , palette='rainbow')
```
