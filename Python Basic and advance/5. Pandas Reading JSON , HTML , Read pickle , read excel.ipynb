## Reading JSON

```python
import pandas as pd
```

```python
data = '{"Employee_name": "john doe" , "branche" : "Computer" , "email":"john@gmail.com" , "job_profile" :[{"title1":"team lead" , "title2":"sr.developer"}]}'
df = pd.read_json(data)
df
```

```python
df.to_json()
```

```python
df.to_json(orient='records')   # orient will simply make the key value pair of column and its rows
```

## Reading HTML 

```python
url = "https://www.fdic.gov/resources/resolutions/bank-failures/failed-bank-list/"
df_html = pd.read_html(url)
df_html[0]
```

```python
url_mcc = "https://en.wikipedia.org/wiki/Mobile_country_code"
df_mcc = pd.read_html(url_mcc )
df_mcc[0]
```

```python
df_mcc = pd.read_html(url_mcc,match="Country" , header=0)
df_mcc[0]
```
