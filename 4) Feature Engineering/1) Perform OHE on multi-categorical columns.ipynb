```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv("mercedese_train.csv" , usecols=['X1', 'X2', 'X3', 'X4', 'X5', 'X6'])
df.head()
```

```python
# lets have look at how many labels are present in each category
for col in df.columns:
    print(col,': has',len(df[col].unique()),'columns')
```

```python
df_OHE = pd.get_dummies(df,drop_first=True)
df_OHE
```

### As here , we took only 5 columns and applied OHE , then it resulted into 117 columns . So imaginge if in real world we have more than 500 categories inside a single column.Then it wont be feasible . So to overcome this we are using approch of most frequent categories.


```python
# lets find the most frequent labels in certain columns (let column = X2)
df['X2'].value_counts().sort_values(ascending = False).head(10)
```

### So above are the most frequent categories in column X2

```python
top_10_labels = [x for x in df['X2'].value_counts().sort_values(ascending = False).head(10).index]
top_10_labels
```

```python
for label in top_10_labels:
    df[label] = np.where(df['X2'] == label, 1, 0)

df[['X2']+top_10_labels].head(10)
```

```python
def one_hot_top(df , variable , top_10_labels):
    # function to create the dummy variables for most frequent labels inside the column
    # we can vary the number of most frequent labels
    
    for label in top_10_labels:
        df[variable+'_'+label] = np.where(df[variable] == label ,1 , 0 )
        
# read the data again
df = pd.read_csv("mercedese_train.csv",usecols=['X1','X2','X3','X4','X5','X6'])
one_hot_top(df,'X2',top_10_labels)
df.head()
```

```python
top_10 = [x for x in df['X1'].value_counts().sort_values(ascending = False).head(10).index]

one_hot_top(df,'X1',top_10)
```

```python
df.head()
```

```python
top_10 = [x for x in df['X3'].value_counts().sort_values(ascending = False).head(10).index]

one_hot_top(df,'X3',top_10)
df.head()
```

```python
top_10 = [x for x in df['X4'].value_counts().sort_values(ascending = False).head(10).index]

one_hot_top(df,'X4',top_10)
top_10 = [x for x in df['X5'].value_counts().sort_values(ascending = False).head(10).index]

one_hot_top(df,'X5',top_10)
```

## Here the OHE columns are now 47 , way less than generated by keeping all the categories from each columns (117)

### Advantages :
- Straight forward to implement
- Does not expand massively the feature space

### Disadvantages:
- Does not add any information that adds variable value more
- Does not keep the information regarding ignored labeled data
