### For handling categorical features we are simply assinging numbers to those features in order

### Like we have described above , we have to give the ranks to each category

### <font color='red'>1) ordinal number encoding </font>

```python
import datetime
```

```python
# take todays date
today_date = datetime.datetime.today()
today_date
```

```python
# we are using timedelta . Using that we are deriving dates
date_list = [today_date - datetime.timedelta(days = x) for x in range(0,20)]
date_list 
```

```python
# we are adding the list to dataframe and giving new column name
import pandas as pd
df = pd.DataFrame(date_list)
df.columns = ['Date']
df
```

```python
df['weekday'] = df['Date'].dt.day_name()
df
```

```python
# we are creating dictionary of weekday names and assigning numbers to each day name
weekday_number = {'Monday' : 1 , 'Tuesday' : 2 , 'Wednesday' : 3 , 
                 'Thursday' : 4 , 'Friday' : 5 , 'Saturday' : 6 , 
                 'Sunday' : 7}
```

```python
df['weekday_number_column']=df['weekday'].map(weekday_number)
df
```

###  <font color='red'>2) Count / frequency encoding </font>

```python
train_set = pd.read_csv('adult.csv' , index_col=None)
train_set.head()
```

```python
# it gives all the unique categories present in column
train_set['workclass'].unique()
```

```python
train_set_cp = train_set.copy()
```

```python
train_set_cp.drop(['age' , 'fnlwgt' , 'educational-num' , 'capital-gain' , 'capital-loss','hours-per-week' , 'income'] ,axis=1, inplace=True)
```

```python
# it gives count of uninque categories
train_set['workclass'].nunique()
```

```python
# here I want to take out how many categories each categorical column contains
for feature in train_set_cp.columns:
    print(feature + ' : ' + str(train_set_cp[feature].nunique()) + ' categories')
```

```python
train_set_cp['native-country'].value_counts()
```

###  we can simply replace '?' with 
- a) make it as  column named 'other' 
- b) simply replace with most frequently occuring category 
- c) simply randomly put any country where '?' is present 

```python
# here we want to replace all the country names with the number of times it occurs 
country_map = train_set_cp['native-country'].value_counts().to_dict()
country_map
```

```python
train_set_cp['native-country'] = train_set_cp['native-country'].map(country_map)
train_set_cp.head()
```

### Advantages:
1) easy to implement
2) feature space will not increase

### Disadvantages :
1) if 2 categories have same frequency then it will assign the same count . so model will not distinguishing that category

###  <font color='red'>3) Target guided ordinal encoding </font>
- here we are encoding one of the categorical column w.r.t output variable i.e Survived
- So based on output variable we are ranking the categorical column values

```python
df = pd.read_csv('titanic-train.csv' , usecols=['Survived','Cabin'])
df.head()
```

```python
# simply replacing cabin NaN values with other label (i.e here say 'missing')
df['Cabin'].fillna('missing', inplace=True)
df.head()
```

```python
# here now we only want the first letter of categorical label from cabin column
df['Cabin'] = df['Cabin'].str[0]
```

```python
df.head()
```

```python
# now we are trying to find out , how many percentage of peaple are survived w.r.t perticular category
(df.groupby(['Cabin'])['Survived'].mean().sort_values())*100
```

```python
ordinal_labels = df.groupby(['Cabin'])['Survived'].mean().sort_values().index
ordinal_labels
```

```python
# now we have to map all above categories to ranks , so we are using enumerate()
ordinal_labels2 = {k:i for i , k in enumerate(ordinal_labels , 1)}  # enumerate is used to obtain indexed list
ordinal_labels2
```

```python
df['cabin_ordinal_enocde'] = df['Cabin'].map(ordinal_labels2)
df.head(20)
```

### <font color='red'>4) Target guided Mean Encoding </font>
- In this we are simply replacing the categories with the mean values based on target feature


```python
df.groupby(['Cabin'])['Survived'].mean()
```

```python
mean_cat = df.groupby(['Cabin'])['Survived'].mean().to_dict()
mean_cat
```

```python
df['Cabin_mean_encode'] = df['Cabin'].map(mean_cat)
df.head()
```

###  Advantages:
- it increases the monotonic relationship between the indpendent and target feature

### Disadvantages :
- It lead to overfitting sometimes
