### in ordinal data , order of the data matters a lot

```python
import pandas as pd 
import datetime  as dt
```

```python
# create todays date 
df_base = dt.datetime.today()
```

```python
# create 20 days back dates nd put in a list
df_date_list = [df_base - dt.timedelta(x) for x in range(0,20)]  # timedelta give sthe difference between 2 dates
df_pd = pd.DataFrame(df_date_list)
df_pd.columns  = ['Date']
df_pd
```

```python
df_pd['day_of_date'] = df_pd['Date'].dt.day_name()
df_pd
```

```python
# engineer categorical variable to perticula labels

weekdays_map = {'Monday' : 1 ,
                'Tuesday' : 2 ,
                'Wednesday' : 3 ,
                'Thursday' : 4 ,
                'Friday' : 5 , 
                'Saturday' : 6 ,
                'Sunday' : 7
               }
```

```python
df_pd['ord_day'] = df_pd.day_of_date.map(weekdays_map)
df_pd.head()
```

### it is suitable for fixed number of categories present in columns , but not for columns having many number of categories
