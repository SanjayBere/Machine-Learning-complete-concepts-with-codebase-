```python
# load the bouston house price dataset
from sklearn.datasets import load_boston
```

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = load_boston()
```

```python
df
```

```python
dataset = pd.DataFrame(df.data)
dataset.head()
```

```python
dataset.columns = df.feature_names
dataset
```

```python
dataset.shape
```

```python
df.target.shape
```

```python
dataset['Price'] = df.target
dataset.head()
```

```python
# now divide the above dataset into dependent and independent features
X = dataset.iloc[:,:-1] # independent featueres
y = dataset.iloc[:,-1] # dependent feauture
```

```python
X
```

```python
y
```

## <font color='red'> Simple Linear Regression implementation</font>

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score

# creating object of linear regression function
lr_clf = LinearRegression()

# doing cv with cv=5 , returns 5 values in form of list 
mse = cross_val_score(lr_clf , X , y , cv = 5 , n_jobs = -1 , scoring = 'neg_mean_squared_error')
print("mse: -------" , mse)

# taking mean of above mse value list
mse_mean = np.mean(mse)
print("\nmse_mean -----",mse_mean)
```

### The more above value is nearer to zero , model actually performs accurately
### But it may lead to overfitting , keep that in mind

## <font color='red'> Ridge regression implementation</font>

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

rdg = Ridge()
param ={'alpha' : [1e-15 , 1e-10 ,1e-8 , 1e-5 , 1e-3 , 1e-2 , 1e-1 , 1,5, 10, 50 , 100]}
rdg_cv = GridSearchCV(rdg , param_grid=param , scoring='neg_mean_squared_error' , cv=5)
rdg_cv.fit(X , y)
```

```python
print(rdg_cv.best_params_)
print(rdg_cv.best_score_)
```

## <font color = 'red'> Lasso regression implementation</font>

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

lasso = Lasso()
param ={'alpha' : [1e-15 , 1e-10 ,1e-8 , 1e-5 , 1e-3 , 1e-2 , 1e-1 , 1,5, 10, 20,50 ,55, 100]}
lasso_cv = GridSearchCV(lasso , param_grid=param , scoring='neg_mean_squared_error' , cv = 5)
lasso_cv.fit(X , y)
```

```python
print(lasso_cv.best_params_)
print(lasso_cv.best_score_)
```

#### Here alpha value choosen was 0.1 and best score = -34.83
