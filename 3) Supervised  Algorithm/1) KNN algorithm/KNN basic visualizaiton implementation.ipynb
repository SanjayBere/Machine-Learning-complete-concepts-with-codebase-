```python
import matplotlib.pyplot as plt
import pandas as pd 
from sklearn import datasets , neighbors
from mlxtend.plotting import plot_decision_regions
```

```python

# core function to find the knn
def knn(df,k):
    x = df[["X","Y"]].values  # independent variable
    y = df["class"].astype(int).values # dependent variable
    clf = neighbors.KNeighborsClassifier(n_neighbors=k)
    clf.fit(x,y)
    
    plot_decision_regions(x,y,clf)
    
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title("Knn implementation with k = "+str(k))
    plt.show()
```

### ushape data

```python
# we are taking k values for say as [1,2,4,10,40,50]
# reading the data file using pandas
df = pd.read_csv("ushape.csv")
for k in [1,2,4,10,40,50]:
    knn(df,k)
```

###  Concentriccircle 

```python
df = pd.read_csv("concertriccir2.csv")
for k in [1,2,4,5,6,10,40,50]:
    knn(df,k)
```
